```python
df_google1 = pd.read_csv('/content/googleads-performance.csv')
print("df 1 shape : ",df_google1.shape)
df_google2 = pd.read_csv('/content/googlead-performance2.csv')
print("df 2 shape : ",df_google1.shape)
result_google = pd.concat([df_google1, df_google2], axis=0, ignore_index=True)
print("final shape for google ads : \n",result_google.shape)
result_google.sample(4)
```

```python
df_ms1 = pd.read_csv('/content/microsoftads-performance.csv')
print("df 1 shape : ",df_ms1.shape)
df_ms2 = pd.read_csv('/content/microsoftads-performance2.csv')
print("df 2 shape : ",df_ms1.shape)
result_ms = pd.concat([df_ms1, df_ms2], axis=0, ignore_index=True)
print("final shape for ms ads : \n",result_ms.shape)
result_ms.sample(4)
```

```python
result_ms['Clicks'] = result_ms['Clicks'].astype(int)
result_ms['Impressions'] = result_ms['Impressions'].astype(int)
result_ms.sample(4)
```

```python
result_ms['Date'] = result_ms['Date'].astype('datetime64[ns]')
result_ms.sample(4)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
ohe = OneHotEncoder(sparse_output=False)  # sparse_output=False ensures output is a dense array
encoded_data = ohe.fit_transform(result_ms[['Campaign type']])

# Convert the result back to a DataFrame with meaningful column names
encoded_df = pd.DataFrame(encoded_data, columns=ohe.get_feature_names_out(['Campaign type']))
print("\nOne-Hot Encoded DataFrame:")
print(encoded_df)
```

```python
df_final = pd.concat([result_ms, encoded_df], axis=1)
df_final.sample(4)
```
